```python
 # Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# importing the statistics module 
from statistics import stdev 
  
# importing frations as parameter values 
from fractions import Fraction as fr 
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "./Resources/mouse_drug_data.csv"
clinical_trial_data_to_load = "./Resources/clinicaltrial_data.csv"
df_mouse_data = pd.read_csv(mouse_drug_data_to_load,encoding="ISO-8859-1")
df_clinical_data = pd.read_csv(clinical_trial_data_to_load,encoding="ISO-8859-1")


```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse_data.head()
```

```python
df_clinical_data.head()
```

```python
# Combine the data into a single dataset
merge_table = pd.merge(df_mouse_data, df_clinical_data, on=["Mouse ID","Mouse ID"])

#Merge table
merge_table.head()
```

# Tumor volume changes over time

```python
#Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = merge_table.groupby(['Timepoint','Drug'], as_index=False)['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tumor_volume = pd.DataFrame(mean_tumor_volume)

# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
error_tumor_volumes = merge_table.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
error_tumor_volumes = pd.DataFrame(error_tumor_volumes)

# Preview DataFrame
error_tumor_volumes.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_volume = mean_tumor_volume.reset_index()
mean_tumor_volume_pvot = mean_tumor_volume.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

error_tumor_volumes = error_tumor_volumes.reset_index()
error_tumor_volumes_pvot = error_tumor_volumes.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

```

```python
# Preview that Reformatting worked
mean_tumor_volume_pvot.head()
```

```python
#plotear cada tratamiento
# Generate the Plot (with Error Bars)
plt.errorbar(mean_tumor_volume_pvot.index, 
             mean_tumor_volume_pvot["Capomulin"], 
             yerr=error_tumor_volumes_pvot["Capomulin"], 
             color="r", 
             marker="o", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(mean_tumor_volume_pvot.index, 
             mean_tumor_volume_pvot["Infubinol"], 
             yerr=error_tumor_volumes_pvot["Infubinol"], 
             color="b", 
             marker="^", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(mean_tumor_volume_pvot.index, 
             mean_tumor_volume_pvot["Ketapril"], 
             yerr=error_tumor_volumes_pvot["Ketapril"], 
             color="g", 
             marker="s", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(mean_tumor_volume_pvot.index, 
             mean_tumor_volume_pvot["Placebo"], 
             yerr=error_tumor_volumes_pvot["Placebo"], 
             color="k", 
             marker="d", 
             linestyle="dashed", 
             linewidth=0.50)

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)


# Save an image of the chart and print it to the screen
plt.savefig("./Tumor_changes_over_time_HWK.png")
```

# Survival Rate


```python
#Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

count_mice = merge_table.groupby(['Timepoint','Drug'], as_index=False)['Mouse ID'].count()

# Convert to DataFrame
count_mice = pd.DataFrame(count_mice)


# Preview DataFrame
count_mice.head()
```

```python
merge_table.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mice = count_mice.reset_index()
count_mice_pvot = count_mice.pivot(index="Timepoint", columns="Drug")["Mouse ID"]

# Preview the Data Frame
count_mice_pvot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(count_mice_pvot.index, 
             (count_mice_pvot["Capomulin"]/25)*100, 
             yerr=count_mice_pvot["Capomulin"], 
             color="r", 
             marker="o", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(count_mice_pvot.index, 
             (count_mice_pvot["Infubinol"]/25)*100, 
             yerr=count_mice_pvot["Infubinol"], 
             color="b", 
             marker="^", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(count_mice_pvot.index, 
             (count_mice_pvot["Ketapril"]/25)*100, 
             yerr=count_mice_pvot["Ketapril"], 
             color="g", 
             marker="s", 
             linestyle="dashed", 
             linewidth=0.50)

plt.errorbar(count_mice_pvot.index, 
             (count_mice_pvot["Placebo"]/25)*100, 
             yerr=count_mice_pvot["Placebo"], 
             color="k", 
             marker="d", 
             linestyle="dashed", 
             linewidth=0.50)

plt.title("Survival rate")
plt.ylabel("Number of mice")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)


# Save the image
plt.savefig("./Survivial_rate_HWK.png")

# Show the Figure
plt.show()
```
